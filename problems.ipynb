Largest Palindrome Product

```python

def solve():
    largest = 0

    for i in range(999,99,-1):
        if i * 999 < largest:
            break

        for j in range(i,99,-1):
            v = i * j
            if v < largest:
                break

            if str(v) == str(v)[::-1]:
                largest = v
                break

    return largest
solve()
```

Smallest Multiple

```python
def gcd(a,b):

    if b == 0:
        return a
    
    return gcd(b,a % b)

def lcm(a,b):
        return a * b // gcd(a,b)

def solve(n):
    result = 1

    for i in range(1,n+1):
        result = lcm(result,i)

    return result

solve(20)
```

Sum Square Difference

```python
def solve(a,b):
    square = 0

    total = 0
    for i in range(a,b):
        square += i * i
        total += i

    square_sum = (total * total)

    return square_sum - square

solve(1,101)
    
```

10 001st Prime

```python
def n_th_prime(n):
    limit = 200000
    sieve = [True] * limit
    sieve[0] = sieve[1] = False

    primes = []
    for i in range(2,limit):
        if sieve[i]:
            primes.append(i)
            if len(primes) == n:
                return i
            
            for j in range(i * i ,limit,i):
                sieve[j] = False
    return -1
    
n_th_prime(10001)
```
